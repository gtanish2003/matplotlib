```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Colored ScatterPlots

```python
iris=pd.read_csv('iris.csv')
```

```python
iris.sample()
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'])
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.show()
```

```python
iris['Species']=iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
```

```python
iris.head()
```

```python
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'])
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.show()
```

```python
# There also exist a hidden parameter called 'cmap' which is used to provide color themes like winter, summer, etc.
# Another parameter is colorbar which is used to show the colorbar on the right side of the plot.
# Another parameter is alpha which is used to set the opapaqueness of the plot.
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='summer',alpha=0.8)
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.colorbar()
plt.show()
```

```python
# How to increase the size of the plot ?
```

```python
plt.figure(figsize=(10,5))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='summer',alpha=0.8)
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.colorbar()
plt.show()
```

# Annotations

```python
batters=pd.read_csv('batter.csv')
```

```python
batters.head()
```

```python
batters.shape
```

```python
sample_df=batters.head(100).sample(25,random_state=5)
```

```python
sample_df.head()
```

```python
plt.scatter(sample_df['avg'],sample_df['strike_rate'])
```

# for writing the text on the plot so that it can be easily understandable by all we use the function plt.text(x,y,'text',fontdict={'size':10,'color':'red'})

```python
plt.figure(figsize=(20,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

# Horizontal lines and Vertical lines

```python
plt.figure(figsize=(20,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
plt.axhline(130,color='red')
plt.axvline(30,color='red')

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

# Subplots

```python
# another way to create the plot in matplotlib
fig,ax=plt.subplots()
ax=plt.scatter(sample_df['avg'],sample_df['strike_rate'])
ax.set_label('Average')
```

```python
fig,ax=plt.subplots(nrows=2,ncols=1,sharex=True)
ax[0].scatter(sample_df['avg'],sample_df['strike_rate'])
ax[1].scatter(sample_df['avg'],sample_df['runs'])
ax[0].set_title('Average vs Strike Rate')
ax[1].set_title('Average vs Runs')
ax[0].set_ylabel('Strike Rate')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Average')
```

# 3D LINE PLOT

```python
x=[0,1,5]
y=[0,10,14]
z=[0,2,15]

fig = plt.figure()
ax=plt.subplot(projection='3d')
ax.scatter3D(x,y,z,s=[100,200,300],c=['red','green','blue'])
ax.plot3D(x,y,z,color='black')
```

# 3d Surface plot

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
```

```python
xx,yy=np.meshgrid(x,y)
```

```python
xx
```

```python
xx.shape
```

```python
yy
```

```python
yy.shape
```

```python
z=xx**2+yy**2
```

```python
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='winter',alpha=0.8)
fig.colorbar(p)
```

# 3d contour plot

### it is used to plot the 3d plot into 2d plot

```python
fig=plt.figure(figsize=(12,8))
ax=plt.subplot()
p=ax.contour(xx,yy,z,cmap='winter')
fig.colorbar(p)
```

```python
fig=plt.figure(figsize=(12,8))
ax=plt.subplot()
p=ax.contourf(xx,yy,z,cmap='winter')
fig.colorbar(p)
```

# Heatmap

```python
delivery=pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
```

```python
delivery.head()
```

```python
temp_df=delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]
```

```python
grid=temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```

```python
plt.figure(figsize=(20,9))
plt.imshow(grid,cmap='viridis')
plt.colorbar()
```

```python
plt.figure(figsize=(20,9))
plt.imshow(grid,cmap='viridis')
plt.yticks(delivery['overs'].unique(),list(range(1,21)))
plt.xticks(np.arange(0,6),list(range(1,7)))
plt.colorbar()
```
