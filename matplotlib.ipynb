```python
batsman=pd.read_csv('sharma-kohli.csv')
```

```python
# plotting multiple objects
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# lables titles
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('Rohit Sharma vs Virat Kohli ')
```

```python
# adding xlabel and ylabel
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('Rohit Sharma vs Virat Kohli ')
plt.xlabel('Year')
plt.ylabel('Runs')
```

```python
# colors(hex) and line (width and style) and marker (size)
plt.plot(batsman['index'],batsman['V Kohli'],color='yellow',linestyle='dashdot',linewidth=5)
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='dashed',linewidth=3)
plt.title('Rohit Sharma vs Virat Kohli ')
plt.xlabel('Year')
plt.ylabel('Runs')
# four line styles are there ---> solid , dashed, dotted , dashdot
```

```python
#  2d plot ke points ko dekhna
#  which is also known as markers
plt.plot(batsman['index'],batsman['V Kohli'],color='yellow',linestyle='dashdot',linewidth=5,marker='D')
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='dashed',linewidth=3)
plt.title('Rohit Sharma vs Virat Kohli ')
plt.xlabel('Year')
plt.ylabel('Runs')
# different markers are ---> D , O , +
# We can also control the size of the markers by using the parameters known as markersize
```

```python
# legends
plt.plot(batsman['index'],batsman['V Kohli'],color='yellow',label='Run Machine')
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',label='Hitman')
plt.title('Rohit Sharma vs Virat Kohli ')
plt.xlabel('Year')
plt.ylabel('Runs')
plt.legend(loc='upper right')
# legend has a parmeter known as loc .. Its default value is best
```

```python
# limiting axis
price=[48000,54000,57000,49000,47000,450]
year=[2015,2016,2017,2018,2019,2020]
plt.plot(year,price)
plt.ylim(0,100000)
plt.xlim(2017,2018)
```

```python
# grid
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('Rohit Sharma vs Virat Kohli ')
plt.xlabel('Year')
plt.ylabel('Runs')
plt.grid()
```

```python
# plt.show() is used to display the plot
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('Rohit Sharma vs Virat Kohli ')
plt.xlabel('Year')
plt.ylabel('Runs')
plt.show()
```

```python
children=[10,20,40,10,30]
colors=['red','blue','green','yellow',
       'pink']
```

```python
# Horizonal Bar Chart
plt.barh(colors, children,color='Black')
```

### If entries are exceeding its limit value . It means that if we have to plot horizontal bar chart then we have to use barh() function. It generally means that the no of columns areincreasing.
### else we have to use bar() function 

```python
#   Multiple.bar plot
df=pd.read_csv ('batsman_season_record.csv')
```

```python
df.head(10)
```

```python
df.shape
```

```python
plt.bar(df['batsman'],df['2015'],width=0.5)
plt.bar(df['batsman'],df['2016'])
plt.bar(df['batsman'],df['2017'])
```

```python
# to overcome the above limitations of multiple plot we do the following:
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)

plt.xticks(np.arange(df.shape[0]),df['batsman'])
```

```python
# label overlapping
## on ccase of overlapping we can use rotation
children=[10,20,40,10,30]
colors=['red red red red red','blue blue blue blue blue','green green green green green green ','yellow yellow yellow yellow yellow yellow',
       'pink pink pink pink pink pink ']
plt.bar(colors, children,color='black')
plt.xticks(rotation='vertical')
 
```

```python
df
```

```python
# stacked bar chart
plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=df['2016']+df['2015'],label='2017')
plt.legend()
plt.show()
```

```python
import pandas as pd

# Create a DataFrame with some random data
df = pd.DataFrame({'values': np.random.randn(1000)})

# Create a histogram with specified bins
df.hist(column='values', bins=20, color='green', alpha=0.7)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
data = np.random.randn(1000)

# Create a histogram with specified bins
plt.hist(data, bins=20, color='blue', alpha=0.7)

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram with Bins')

# Show the plot
plt.show()
```

# Pie chart

### Univariate and Bivariate Analysis
###  Categorical vs Numerical
### Use case - to find contribution on a standard scale


```python
data=[100,56,100,45]
subjects=['Maths','English','Science','Social']
plt.pie(data,labels=subjects)
plt.show()
```

```python
df=pd.read_csv('gayle-175.csv' )
```

```python
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'])
```

```python
# Therer is also one more parameter known as autopct to show the percentage if each contribution
df=pd.read_csv('gayle-175.csv' )
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
```

```python
# colors speicification
df=pd.read_csv('gayle-175.csv' )
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['red','green','blue','yellow','black'])
```

```python
df.size
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],explode=[0.1,0,0,0,0,0])
plt.show()
```

```python
# another parameter is shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],explode=[0.1,0,0,0,0,0],shadow=True)
```
